```python
import os
from langchain_openai import OpenAI
from langchain import PromptTemplate, LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, AIMessage
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')

template_a = PromptTemplate(
    input_variables=["input"],
    template="You are in a debate. You will be speaking FOR the topic. Respond to: {input}"
)
template_b = PromptTemplate(
    input_variables=["input"],
    template="You are in a debate. You will be speaking AGAINST the topic. Respond to: {input}"
)

closing_template_a = PromptTemplate(
    input_variables=["input"],
    template="You are in a debate. You will be speaking FOR the topic. Provide your closing arguments based on the debate: {input}"
)
closing_template_b = PromptTemplate(
    input_variables=["input"],
    template="You are in a debate. You will be speaking AGAINST the topic. Provide your closing arguments based on the debate: {input}"
)

llm_a = ChatOpenAI(api_key=api_key, model_name='gpt-3.5-turbo')
llm_b = ChatOpenAI(api_key=api_key, model_name='gpt-3.5-turbo')

chain_a = LLMChain(llm=llm_a, prompt=template_a)
chain_b = LLMChain(llm=llm_b, prompt=template_b)
closing_chain_a = LLMChain(llm=llm_a, prompt=closing_template_a)
closing_chain_b = LLMChain(llm=llm_b, prompt=closing_template_b)

input_text = input("Enter the initial debate topic: ")
max_rounds = 10
round_counter = 0

while round_counter < max_rounds:
    response_a = chain_a.invoke({"input": input_text})
    response_a_text = response_a['text'] 
    print(f"Debater A (FOR): {response_a_text}")

    response_b = chain_b.invoke({"input": response_a_text})
    response_b_text = response_b['text']
    print(f"Debater B (AGAINST): {response_b_text}")

    round_counter += 1  

    if round_counter % 2 == 0 and round_counter < max_rounds:
        input_text = input("Enter a new debate topic or continue the current one (or type 'exit' to quit): ")
        if input_text.lower() == 'exit':
            break
    else:
        input_text = response_b_text

closing_input = response_b_text
closing_response_a = closing_chain_a.invoke({"input": closing_input})
closing_response_b = closing_chain_b.invoke({"input": closing_input})

print("\nClosing Arguments:")
print(f"Debater A (FOR): {closing_response_a['text']}")
print(f"Debater B (AGAINST): {closing_response_b['text']}")
```
